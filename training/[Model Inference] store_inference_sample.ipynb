```python
# Import necessary packages
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
# Load the model
model = keras.models.load_model('/content/drive/MyDrive/mynt/store_sale_predictor')
```

```python
# Enter the sample input accordingly: Store	DayOfWeek	Customers	Open	Promo	StateHoliday	SchoolHoliday	Month	Day
store = 1
day_of_week = 4
customers = 650
open = 1
promo = 1
state_holiday = 0
school_holiday = 0
month = 4
day = 30
```

```python
# Transform the inputs to np array
new_input = np.array([store, 
                      day_of_week, 
                      customers, 
                      open, 
                      promo, 
                      state_holiday, 
                      school_holiday, 
                      month, 
                      day]).reshape(1, -1)
```

```python
# Reshape to (batch_size, timesteps, input_dim)
new_input = np.reshape(new_input, (new_input.shape[0], 1, new_input.shape[1]))
```

```python
# Inference the input
prediction = model.predict(new_input)
```
